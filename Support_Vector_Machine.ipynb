```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(10)
```

```python
df['target'] = iris.target
df.head(10)
```

```python
len(iris.data)
```

```python
# df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])

for i in range(len(iris.data)):
    df['flower_name'] = iris.target_names[iris.target]

    
df[45:105]
```

```python
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
X.head(10)
```

```python
y = df.target
y.head(10)
```

```python
df0 = df[0:50]
df1 = df[50:100]
df2 = df[100:150]
```

```python
df0
```

```python
df1
```

```python
df2
```

<h3><b>Sepal Length vs Sepal Width (Setosa vs Virginica)</b></h3>

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='blue', marker='+')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Setosa vs Virginica')
```

<h3><b>Petal Length vs Petal Width (Versicolor vs Virginica)</b></h3>

```python
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='green', marker='+')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='blue', marker='+')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Versicolor vs Virginica')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

```python
model.score(X_test, y_test)
```

<h3><b>Tune Parameters</b></h3>

<h2><b>1. Regularization (C)</b></h2>

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
```

```python
model_C.score(X_test, y_test)
```

```python
model_C = SVC(C=50)
model_C.fit(X_train, y_train)
```

```python
model_C.score(X_test, y_test)
```

<h2><b>2. Gamma</b></h2>

```python
model_g = SVC(gamma=5)
model_g.fit(X_train, y_train)
```

```python
model_g.score(X_test, y_test)
```

<h2><b>3. Kernel</b></h2>

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
